## 2. Data Types: Series & DataFrame

Pandas has two core data structures you'll work with constantly.

### 2.1 Series

A **Series** is a one-dimensional labeled array. Think of it as a single column in a spreadsheet — each value has an index (label) attached to it.

**Syntax:** `pd.Series(data, index=None)`

````python
# Creating a basic Series from a list
series = pd.Series(["BMW", "Toyota", "Honda"])
series
````

````python
# Another Series example
colors = pd.Series(["Red", "Blue", "White"])
colors
````

````python
# Series with a custom index
prices = pd.Series([4000, 5000, 7000], index=["car_a", "car_b", "car_c"])
prices
````

### 2.2 DataFrame

A **DataFrame** is a two-dimensional, labeled data structure made of rows and columns — similar to a spreadsheet or SQL table, but programmable.

You can build a DataFrame from a dictionary of Series or lists.

````python
# Creating a DataFrame from two Series
car_makes = pd.Series(["BMW", "Toyota", "Honda"])
car_colors = pd.Series(["Red", "Blue", "White"])

car_data = pd.DataFrame({
    "Car_Manufactured": car_makes,
    "Colours": car_colors
})
car_data
````

````python
# Creating a DataFrame directly from a dictionary of lists
sample_df = pd.DataFrame({
    "Make": ["Toyota", "Honda", "BMW"],
    "Year": [2018, 2020, 2022],
    "Price": [15000, 20000, 45000]
})
sample_df
````

> **Note:** Writing DataFrames manually can be tedious for large datasets. In practice, we almost always **import** data from files.

## 3. Importing & Exporting Data

Pandas supports reading from and writing to many file formats: CSV, Excel, JSON, SQL, and more.

### 3.1 Reading a CSV File

`pd.read_csv("filename.csv")` reads a comma-separated values file into a DataFrame.

````python
# Import car sales data from a CSV file
car_sales = pd.read_csv("car-sales.csv")
car_sales
````

### 3.2 Exporting a DataFrame to CSV

`df.to_csv("filename.csv", index=False)` saves the DataFrame to a CSV file.
Setting `index=False` prevents the row numbers from being written as an extra column.

````python
# Export to CSV (index=False avoids saving the default integer index)
car_sales.to_csv("exported-car-sales.csv", index=False)
print("File exported successfully!")
````

````python
# Read the exported file back to verify it looks correct
exported_car_sales = pd.read_csv("exported-car-sales.csv")
exported_car_sales
````

### 3.3 Other Common File Formats

| Format | Read | Write |
|--------|------|-------|
| CSV | `pd.read_csv()` | `df.to_csv()` |
| Excel | `pd.read_excel()` | `df.to_excel()` |
| JSON | `pd.read_json()` | `df.to_json()` |
| SQL | `pd.read_sql()` | `df.to_sql()` |

```python
# Examples (not run — requires the respective files/databases)
df = pd.read_excel("data.xlsx", sheet_name="Sheet1")
df = pd.read_json("data.json")
```

## 5. Viewing & Selecting Data

Pandas gives you many flexible ways to look at and select subsets of your data.

### 5.1 head() and tail()

- `df.head(n)` — returns the first n rows (default: 5)
- `df.tail(n)` — returns the last n rows (default: 5)

````python
# First 5 rows (default)
car_sales.head()
````

````python
# First 7 rows
car_sales.head(7)
````

````python
# Last 5 rows (default)
car_sales.tail()
````

### 5.2 .loc — Label-Based Selection

`.loc[]` selects data by **label** (the index value).

**Key rule:** `.loc` uses the index *label*, not the position.

````python
# Create a Series with non-default (non-sequential) index
animals = pd.Series(
    ["cat", "dog", "lion", "tiger", "pandas"],
    index=[0, 3, 9, 8, 3]  # Note: index 3 appears twice!
)
animals
````

````python
# .loc[3] returns ALL rows with label 3
animals.loc[3]
````

````python
# .loc[9] returns the single row with label 9
animals.loc[9]
````

````python
# .loc on a DataFrame — returns the row with label 3
car_sales.loc[3]
````

### 5.3 .iloc — Position-Based Selection

`.iloc[]` selects data by **integer position** (like Python list indexing).

**Key difference:** `.iloc` always uses 0-based position, regardless of the index labels.

````python
# .iloc[3] returns the item at POSITION 3 (0-indexed) — which is 'tiger'
animals.iloc[3]
````

````python
# Slice: positions 0, 1, 2 (first 3 items)
animals.iloc[:3]
````

````python
# DataFrame: row at position 3
car_sales.iloc[3]
````

````python
# DataFrame: first 3 rows (positions 0, 1, 2)
car_sales.iloc[:3]
````

> **loc vs iloc summary:**
> - `.loc[3]` → row(s) where index **label** == 3
> - `.iloc[3]` → row at **position** 3 (always 0-indexed)
> 
> When the index is the default integer range, they often return the same row — but they behave differently if the index has been customised.

### 5.4 Column Selection

You can select a single column in two ways:
- `df["column_name"]` — bracket notation (always works)
- `df.column_name` — attribute notation (only works when the name has no spaces or special characters)

````python
# Bracket notation (recommended — works with any column name)
car_sales["Make"]
````

````python
# Dot notation (convenient, but limited to simple names)
car_sales.Make
````

````python
car_sales["Colour"]
````

### 5.5 Boolean Filtering (Conditional Selection)

You can filter rows by passing a boolean condition inside `[]`.

````python
# Rows where Make is Toyota
car_sales[car_sales["Make"] == "Toyota"]
````

````python
# Rows where Odometer > 100,000 km
car_sales[car_sales["Odometer (KM)"] > 100000]
````

### 5.6 groupby() and crosstab()

These tools help you summarise and cross-tabulate data.

````python
# Cross-tabulation: count of each Make by number of Doors
pd.crosstab(car_sales["Make"], car_sales["Doors"])
````

````python
# Group by Make and compute mean of numeric columns
car_sales.groupby("Make").mean(numeric_only=True)
````

````python
# Combining filtering + groupby:
# Average odometer and doors for white Toyotas only
car_sales[
    (car_sales["Colour"] == "White") &
    (car_sales["Make"] == "Toyota")
].groupby("Make").mean(numeric_only=True)
````

### 5.7 Basic Plotting

Pandas has built-in plotting via Matplotlib. Use `.plot()` for a line chart and `.hist()` for a histogram.

````python
# Line plot of Odometer values
car_sales["Odometer (KM)"].plot()
````

````python
# Histogram of Odometer values
car_sales["Odometer (KM)"].hist()
````

````python
# First, let's check the Price column type — it may be a string
car_sales["Price"].dtype
````

````python
# Convert Price from string ("$4,000.00") to float (4000.0)
car_sales["Price"] = (
    car_sales["Price"]
    .replace(r"[$,]", "", regex=True)  # remove $ and commas
    .astype(float)
)
car_sales
````

````python
# Line plot of Price values (now numeric)
car_sales["Price"].plot()
````

## 6. Manipulating Data

Pandas makes it easy to transform, clean, and add to your data.

### 6.1 String Operations

Use `.str` accessor to apply string methods element-wise to a column.

````python
# View lowercase version (without changing the column)
car_sales["Make"].str.lower()
````

````python
# Actually apply the change to the column
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
````

### 6.2 Handling Missing Data (NaN)

Pandas represents missing values as `NaN` (Not a Number). It's important to detect and handle them.

````python
# Load a dataset that has missing values
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
````

````python
# Check for missing values in each column
car_sales_missing.isna().sum()
````

````python
# Fill missing Odometer values with the column mean
mean_odometer = car_sales_missing["Odometer"].mean()
print(f"Mean Odometer: {mean_odometer:.2f}")

car_sales_missing["Odometer"].fillna(mean_odometer)
````

````python
# Apply the fill permanently using inplace or assignment
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(
    car_sales_missing["Odometer"].mean()
)
car_sales_missing
````

````python
# Alternatively, drop rows with ANY missing values
car_sales_missing.dropna()
````

### 6.3 Adding Columns

You can add a new column using a Series, a list, a scalar, or a computed value.

````python
# Add a column from a Series (index alignment matters!)
seats_column = pd.Series([5, 5, 5, 5, 5])
car_sales["Seats"] = seats_column
car_sales
````

````python
# Fill any NaN in Seats (caused by index mismatch) with 5
car_sales["Seats"] = car_sales["Seats"].fillna(5)
car_sales
````

````python
# Add a column from a Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel Per 100KM (L)"] = fuel_economy
car_sales
````

````python
# Add a column derived from other columns
car_sales["Total Fuel Used (L)"] = (
    car_sales["Odometer (KM)"] / 100 * car_sales["Fuel Per 100KM (L)"]
)
car_sales
````

````python
# Add a column from a single scalar value (broadcasts to all rows)
car_sales["Number of Wheels"] = 4
car_sales["Passed Road Safety"] = True
car_sales.dtypes
````

### 6.4 Removing Columns

Use `df.drop("column_name", axis=1)` to remove a column.  
- `axis=1` means columns; `axis=0` means rows.
- Use `inplace=True` or reassign to make the change permanent.

````python
# Remove the duplicate fuel column
car_sales = car_sales.drop("Total Fuel Used (L)", axis=1)
car_sales
````

### 6.5 Sampling & Resetting the Index

`.sample(frac=...)` randomly selects a fraction of rows — useful for train/test splits in machine learning.

````python
# Shuffle all rows (sample 100% of the data randomly)
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
````

````python
# Sample only 20% of rows
car_sales_shuffled.sample(frac=0.2)
````

````python
# After shuffling, the index is out of order — reset it to 0, 1, 2, ...
car_sales_shuffled = car_sales_shuffled.reset_index(drop=True)
car_sales_shuffled
````

### 6.6 Applying Functions with .apply()

`.apply(func)` applies any function to each element of a column. It's commonly used with `lambda` for quick transformations.

````python
# Convert Odometer from km to miles (1 km ≈ 0.621 miles)
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
````

````python
# You can also pass a named function instead of a lambda
def km_to_miles(km):
    return km * 0.621371

# Example (not applied to keep data consistent):
# car_sales["Odometer (Miles)"] = car_sales["Odometer (KM)"].apply(km_to_miles)
````
